# Credit Risk & Loan Default Prediction - EDA

This notebook contains exploratory data analysis for the credit risk dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## Load Data

```python
df = pd.read_csv('/Users/lk/Documents/Projects/credit-risk-loan-default-prediction/data/raw/credit_data.csv')
print(f"Dataset shape: {df.shape}")
print(f"\nFirst 5 rows:")
df.head()
```

## Data Overview

```python
print("Data Types:")
print(df.dtypes)
print("\n" + "="*50)
print("\nMissing Values:")
print(df.isnull().sum())
print("\n" + "="*50)
print("\nStatistical Summary:")
df.describe()
```

## Target Variable Distribution

```python
plt.figure(figsize=(8, 5))
df['default'].value_counts().plot(kind='bar', color=['steelblue', 'coral'])
plt.title('Target Variable Distribution')
plt.xlabel('Default (0=No, 1=Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('/Users/lk/Documents/Projects/credit-risk-loan-default-prediction/models/target_distribution.png', dpi=300)
plt.show()

print(f"\nTarget distribution:")
print(df['default'].value_counts(normalize=True))
```

## Feature Distributions

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop('default')
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    ax = axes[i]
    df[col].hist(bins=30, ax=ax, color='steelblue', edgecolor='black')
    ax.set_title(col)
    ax.set_xlabel(col)
    
for j in range(len(numeric_cols), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.savefig('/Users/lk/Documents/Projects/credit-risk-loan-default-prediction/models/feature_distributions.png', dpi=300)
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(14, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('/Users/lk/Documents/Projects/credit-risk-loan-default-prediction/models/correlation_heatmap.png', dpi=300)
plt.show()

print("\nTop Correlations with Target:")
corr_with_target = corr_matrix['default'].drop('default').abs().sort_values(ascending=False)
print(corr_with_target)
```

## Default Rate by Key Features

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Credit Score bins
df['credit_score_bin'] = pd.cut(df['credit_score'], bins=[0, 580, 670, 740, 800, 850], 
                               labels=['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'])
df.groupby('credit_score_bin')['default'].mean().plot(kind='bar', ax=axes[0, 0], color='steelblue')
axes[0, 0].set_title('Default Rate by Credit Score')
axes[0, 0].set_xlabel('Credit Score Category')
axes[0, 0].set_ylabel('Default Rate')
axes[0, 0].tick_params(axis='x', rotation=45)

# Age bins
df['age_bin'] = pd.cut(df['age'], bins=[0, 25, 35, 45, 55, 65, 100], 
                     labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+'])
df.groupby('age_bin')['default'].mean().plot(kind='bar', ax=axes[0, 1], color='coral')
axes[0, 1].set_title('Default Rate by Age Group')
axes[0, 1].set_xlabel('Age Group')
axes[0, 1].set_ylabel('Default Rate')
axes[0, 1].tick_params(axis='x', rotation=45)

# Income bins
df['income_bin'] = pd.cut(df['income'], bins=[0, 30000, 50000, 75000, 100000, float('inf')], 
                         labels=['<30k', '30-50k', '50-75k', '75-100k', '100k+'])
df.groupby('income_bin')['default'].mean().plot(kind='bar', ax=axes[1, 0], color='green')
axes[1, 0].set_title('Default Rate by Income')
axes[1, 0].set_xlabel('Income Category')
axes[1, 0].set_ylabel('Default Rate')
axes[1, 0].tick_params(axis='x', rotation=45)

# Employment years
df['employment_bin'] = pd.cut(df['employment_years'], bins=[-1, 1, 3, 5, 10, float('inf')], 
                            labels=['<1yr', '1-3yr', '3-5yr', '5-10yr', '10yr+'])
df.groupby('employment_bin')['default'].mean().plot(kind='bar', ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Default Rate by Employment Years')
axes[1, 1].set_xlabel('Employment Years')
axes[1, 1].set_ylabel('Default Rate')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('/Users/lk/Documents/Projects/credit-risk-loan-default-prediction/models/default_rate_analysis.png', dpi=300)
plt.show()
```
